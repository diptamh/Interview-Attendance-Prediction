```python
def clean_date(date):
    date = date.str.strip()
    date = date.str.split("&").str[0]
    date = date.str.replace('–', '/')
    date = date.str.replace('.', '/')
    date = date.str.replace('Apr', '04')
    date = date.str.replace('-', '/')
    date = date.str.replace(' ', '/')
    date = date.str.replace('//+', '/')
    return date
df['Date'] = clean_date(df['Date'])
print(df['Date'])
```

# Logistic Regression

```python
X = df[['Permission','Alt_phone', 'Resume_Printout','Clarify_Venue','Shared_Letter','Expected']]
y = df['Attended']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions=logmodel.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Imputer
my_pipeline = make_pipeline(Imputer(), RandomForestRegressor())
from sklearn.model_selection import cross_val_score
scores = cross_val_score(my_pipeline, X, y, scoring='neg_mean_absolute_error')
print(scores)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import metrics
log_reg_matrix = metrics.confusion_matrix(y_test,predictions)
sns.heatmap(log_reg_matrix, annot=True,  fmt='');
title = 'Logistic Regression'
plt.title(title);
```

# Gaussian NB

```python
from sklearn.naive_bayes import GaussianNB
GNBmodel=GaussianNB()
```

```python
GNBmodel.fit(X,y)
nb=GNBmodel.score(X,y)
```

```python
pred1=GNBmodel.predict(X_test)
t1=sum(X==0 for X in pred1-y_test)/len(pred1)
print (t1)
print (pred1)
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
```

```python
from sklearn.metrics import roc_curve,auc
from scipy import interp
from itertools import cycle
n_classes=1
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(np.array(pd.get_dummies(y_test))[:, i], np.array(pd.get_dummies(pred1))[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
```

```python
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += interp(all_fpr, fpr[i], tpr[i])
```

```python
mean_tpr /= n_classes
```

```python
fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
```

```python
lw=2
plt.figure(figsize=(8,5))
plt.plot(fpr["macro"], tpr["macro"],
         label='macro-average ROC curve (area = {0:0.2f})'
               ''.format(roc_auc["macro"]),
         color='green', linestyle=':', linewidth=4)
```

```python
colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
''.format(i, roc_auc[i]))
```

```python
plt.plot([0, 1], [0, 1], 'k--',color='red', lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.annotate('Random Guess',(.5,.48),color='red')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic for Naive Bayes - IRIS DATASET')
plt.legend(loc="lower right")
plt.show()
```
